```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches
```

```python
# Model parameters from Par_file
xmin = 0.0
xmax = 4000.0
nx = 80
# For demonstration, we assume nz = 60 (as suggested by region definitions)
nz = 60

dx = (xmax - xmin) / nx
# Here we assume a total depth; if not specified, choose a convenient value (e.g., 3000 m)
zmax = 3000.0
dz = zmax / nz

# List of region definitions: (nxmin, nxmax, nzmin, nzmax, material)
regions = [
    (1,20, 1,20, 1),
    (1,20,21,40, 2),
    (1,20,41,60, 1),
    (21,40,1,20, 2),
    (21,40,21,40, 1),
    (21,40,41,60, 2),
    (41,60,1,20, 1),
    (41,60,21,40, 2),
    (41,60,41,60, 1),
    (61,80,1,20, 2),
    (61,80,21,40, 1),
    (61,80,41,60, 2)
]

plt.figure(figsize=(8, 6))
ax = plt.gca()

# Create a color mapping for materials (customize as needed)
colors = {1: 'lightblue', 2: 'lightgreen'}

for region in regions:
    ixmin, ixmax, izmin, izmax, mat = region
    # Convert element indices to coordinates (note: indices start at 1)
    x_left = xmin + (ixmin - 1) * dx
    width = (ixmax - ixmin + 1) * dx
    z_top = (izmin - 1) * dz
    height = (izmax - izmin + 1) * dz
    # Draw a rectangle (we use negative z to have the surface at the top)
    rect = patches.Rectangle((x_left, -z_top - height), width, height, 
                             facecolor=colors.get(mat, 'gray'), edgecolor='k', alpha=0.5)
    ax.add_patch(rect)
    # Optionally, label the material number at the center
    plt.text(x_left + width/2, -z_top - height/2, str(mat), 
             ha='center', va='center', fontsize=10)

plt.xlabel('Horizontal distance (m)')
plt.ylabel('Depth (m)')
plt.title('Block Diagram of the Velocity Model')
plt.xlim(xmin, xmax)
plt.ylim(0, -zmax)
plt.gca().invert_yaxis()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 6))

# Plot the source
x_source = 2500.
z_source = 2500.
plt.plot(x_source, -z_source, 'r*', markersize=12, label='Source')

# Receiver Set 1 (horizontal line)
nrec1 = 11
xdeb1 = 300.
xfin1 = 3700.
z_rec1 = 2200.  # constant depth
x_receivers1 = np.linspace(xdeb1, xfin1, nrec1)
z_receivers1 = np.full(nrec1, z_rec1)
plt.plot(x_receivers1, -z_receivers1, 'gs', markersize=8, label='Receivers (Set 1)')

# Receiver Set 2 (vertical line)
nrec2 = 11
xdeb2 = 2500.
zdeb2 = 2500.
zfin2 = 0.
# x coordinate is constant for all receivers in set 2
x_receivers2 = np.full(nrec2, xdeb2)
z_receivers2 = np.linspace(zdeb2, zfin2, nrec2)
plt.plot(x_receivers2, -z_receivers2, 'bo', markersize=8, label='Receivers (Set 2)')

plt.xlabel('Horizontal distance (m)')
plt.ylabel('Depth (m)')
plt.title('Source and Receiver Configuration')
plt.legend()
plt.ylim(0, -zmax)
plt.gca().invert_yaxis()
plt.grid(True)
plt.show()
```
